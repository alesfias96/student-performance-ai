Analizziamo distribuzioni di accuracy, tempi e errori.

```python
import pandas as pd
import matplotlib.pyplot as plt

answers = pd.read_csv("~/Desktop/student-performance-ai/data/raw/student_answers.csv")

# accuracy globale
overall_acc = answers["is_correct"].mean()
print("Overall accuracy:", overall_acc)
```

```python
# Accuracy per topic
topic_acc = answers.groupby("topic")["is_correct"].mean().sort_values(ascending=False)
topic_acc.plot(kind="bar")
plt.ylabel("Accuracy")
plt.title("Accuracy per topic")
plt.show()
```

```python
# Distribuzione tempi
answers["time_seconds"].plot(kind="hist", bins=40)
plt.xlabel("Time (seconds)")
plt.title("Time distribution")
plt.show()
```

```python
# Errori piÃ¹ frequenti (solo sbagliate)
err = answers[answers["is_correct"]==0]["error_type"].value_counts()
err.plot(kind="bar")
plt.title("Error types (wrong answers)")
plt.show()
```
